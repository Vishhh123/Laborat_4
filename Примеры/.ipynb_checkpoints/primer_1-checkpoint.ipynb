```python
import pandas as pd

s = pd.Series([10, 20, 30, 40, 50], index=['a', 'b', 'c', 'd', 'e'])

print(s.iloc[0])
print(s.iloc[2])
print(s.iloc[-1])
```

```python
import pandas as pd

s = pd.Series([10, 25, 8, 30, 15], index=['a', 'b', 'c', 'd', 'e'])

filtered_s = s[s > 10]

print(filtered_s)
```

```python
print(s > 10)
```

```python
filtered_s = s[(s >= 10) & (s <= 30)]
print(filtered_s)
```

```python
filtered_s = s[s.index.isin(['b', 'd'])]
print(filtered_s)
```

```python
s_with_nan = pd.Series([10, None, 8, 30, None], index=['a', 'b', 'c', 'd', 'e'])

filtered_s = s_with_nan[s_with_nan.notnull()]
print(filtered_s)
```

```python
import pandas as pd
import numpy as np

s = pd.Series([10, np.nan, 30, None, 50])

print(s.isnull())
```

```python
print(s.notnull())
```

```python
filtered_s = s[s.notnull()]
print(filtered_s)
```

```python
missing_values = s[s.isnull()]
print(missing_values)
```

```python
s = pd.Series([10, np.nan, 30, None, 50])

s_filled = s.fillna(0)
print(s_filled)
```

```python
s_filled = s.fillna(s.mean())
print(s_filled)
```

```python
s_cleaned = s.dropna()
print(s_cleaned)
```

```python
import pandas as pd

s = pd.Series([10, 20, 30, 40, 50])

s_multiplied = s * 2
print(s_multiplied)

```

```python
s1 = pd.Series([1, 2, 3, 4, 5], index=['a', 'b', 'c', 'd', 'e'])
s2 = pd.Series([10, 20, 30, 40, 50], index=['a', 'b', 'c', 'd', 'e'])

s_sum = s1 + s2
print(s_sum)
```

```python
s3 = pd.Series([100, 200, 300], index=['a', 'b', 'f'])

s_result = s1 + s3
print(s_result)
```

```python
s_result = s1.add(s3, fill_value=0)
print(s_result)
```

```python
s = pd.Series([1, 2, 3, 4, 5])

s_squared = s.apply(lambda x: x ** 2)
print(s_squared)
```

```python
def custom_function(x):
    return f"Value: {x}"

s_transformed = s.apply(custom_function)
print(s_transformed)
```

```python
import numpy as np

s_log = s.apply(np.log)
print(s_log)
```
